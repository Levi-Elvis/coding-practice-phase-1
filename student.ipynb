# Removing missing values

```python
df = df.dropna(axis=0, how='all')  # Drop rows with all NaN values
df = df.dropna(axis=1, how='all')  # Drop columns with all NaN values
df = df.dropna(subset=['Purpose.of.flight'], how='any')  # Drop rows with NaN in 'Purpose.of.flight'
df = df.dropna(subset=['Publication.Date'], how='any')  # Drop rows with NaN in 'Publication.Date'
df = df.dropna(subset=['Aircraft.Category'], how='any')  # Drop rows with NaN in 'Aircraft.Category'
df = df.dropna(subset=['Report.Status'], how='any')  # Drop rows with NaN in 'Report.Status'
df
```

# String Cleaning
Remove extra spaces

```python
df['Purpose.of.flight'] = df['Purpose.of.flight'].str.strip()
df['Purpose.of.flight'] = df['Purpose.of.flight'].str.replace(' ', '_')  # Replace spaces with underscores
df['Aircraft.Category'] = df['Aircraft.Category'].str.strip()
df['Aircraft.Category'] = df['Aircraft.Category'].str.replace(' ', '_')  # Replace spaces with underscores
df['Report.Status'] = df['Report.Status'].str.strip()
```
